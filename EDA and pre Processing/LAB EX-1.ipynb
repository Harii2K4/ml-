```python
df['sex'].value_counts()  #gives the value counts for each recurring value
```

```python
df['age'].unique()  # give an array of all the unique values in the age series
```

```python
#statistical values
print(df['age'].mean())
print(df['age'].max())
print(df['age'].min())
print(df['age'].count())
print(df['age'].median())
print(df['age'].count())
print(df['age'].quantile(0.75))
```

```python
#Dispersion of values
print(df['chol'].std()) # find the standard deviation an column
print(df['chol'].var())  # find the variance of a an column
print(df['chol'].corr(df['age']))  # find the correlation between to columns
```

```python
#crosstabs
print(pd.crosstab(index=df['chol'],columns='age',dropna=True))# to create frequency table and dropna to disregard nan value
print(pd.crosstab(index=df['chol'],columns=df['age']))#two way table
```

```python
print(pd.crosstab(index=df['age'],columns=df['sex'],normalize=True))# TWO WAY -joint probability
print(pd.crosstab(index=df['age'],columns=df['sex'],normalize=True,margins=True)) # TWO WAY -margin probability
```

```python
# two way conditional probability
print(pd.crosstab(index=df['age'],columns=df['sex'],normalize='columns',margins=True))
print(pd.crosstab(index=df['age'],columns=df['sex'],normalize='index',margins=True))
```
