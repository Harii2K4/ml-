```python
import numpy as np

def gradient_descent(x, y, a, epoch):
    count = 0
    b = 0
    w = np.ones((8, 1))  # Initialize w with shape (8, 1)

    for _ in range(epoch):
        # Compute the predictions
        linear = np.dot(x, w) + b

        # Compute the gradients
        dw = np.dot(x.T, (linear - y)) / len(y)
        db = np.mean(linear - y)

        # Update the parameters
        w -= a * dw
        b -= a * db

        # Check for NaN or inf values
        if np.isnan(w).any() or np.isinf(w).any() or np.isnan(b) or np.isinf(b):
            print("NaN or inf value encountered in parameters")
            break

        count += 1

    return w, b

# Example usage
# Assuming x and y are your input features and target values, respectively
x = np.random.rand(3000, 8)  # Example feature matrix
y = np.random.rand(3000, 1)  # Example target values
a = 0.01  # Learning rate
epoch = 1000  # Number of iterations

# Normalize the data (optional but recommended)
x = (x - np.mean(x, axis=0)) / np.std(x, axis=0)
y = (y - np.mean(y)) / np.std(y)

w, b = gradient_descent(x, y, a, epoch)
print("Weights:", w)
print("Bias:", b)




```
